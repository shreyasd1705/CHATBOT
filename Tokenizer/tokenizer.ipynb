```python

import string
fp = open("Sarcasm_tweet_truth.txt", 'r')
id_truth_map = {}
```

```python



for line in fp:
  line = line.strip()
  if line == '':
    continue
  elif line[0] in string.digits:
    current_id = line
  else:
    id_truth_map[current_id] = line
```

```python
import pandas as pd
id_tweet = pd.DataFrame(id_tweet_map.items(), columns=['tweet_id', 'tweet'])
lbl_tweet = pd.DataFrame(id_truth_map.items(), columns=['tweet_id', 'sarcasm'])
dataset = pd.merge(id_tweet, lbl_tweet, on='tweet_id')
dataset = dataset.replace({'sarcasm': {'YES': 1, 'NO': 0}})
```

```python
from transformers import AutoTokenizer
```

```python
llm_hin_eng = pd.read_csv('LLM_models-Hin-Eng.csv')
llm_telugu_eng = pd.read_csv('LLM_models-Telugu-Eng.csv')
llm_tensor_hin_eng = pd.read_csv('LLM_models-Tensor-Hin-Eng.csv')
llm_tensor_telugu_eng = pd.read_csv('LLM_models-Tensor-Telugu-Eng.csv')
```

```python
hf_hin_eng = set(llm_hin_eng['model'])
hf_telugu_eng = set(llm_telugu_eng['Model'])
tensor_hin_eng = set(llm_tensor_hin_eng['Model Name'])
#tensor_telugu_eng = set(llm_tensor_telugu_eng['Model Name'])
```

```python
invalid = []
for i in [5, 16, 26, 27, 50, 51, 52, 53, 55, 60, 61, 64, 65, 67, 71, 72,79, 80, 81,90, 91, 92, 93]:
  model = llm_tensor_hin_eng.iloc[i]['Model Name']
  invalid.append(model)
  print(model)
```

```python
invalid_ids=set(invalid)
all_ids = tensor_hin_eng-hf_hin_eng-hf_telugu_eng
relevant_ids = all_ids - invalid_ids
```

```python




import pandas as pd
from transformers import AutoTokenizer

# Rest of your code
HF_token_analysis = pd.DataFrame()
HF_token_analysis['instance'] = dataset['tweet']

for valid_model in relevant_ids:
    # Remove trailing and leading spaces from the model identifier
    valid_model = valid_model.strip()

    tokenizer = AutoTokenizer.from_pretrained(valid_model)
    data = tokenizer(dataset['tweet'].tolist())
    token_count = []

    for instance in range(0, len(dataset)):
        token_count.append(len(data['input_ids'][instance]) - 2)

    HF_token_analysis[valid_model] = token_count

HF_token_analysis.to_csv('HF_transformer_Hin_Eng_analysis.csv', index=False)

```
